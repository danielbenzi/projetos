Visualizacao de uma tabela com variaçao percentual YoY para o Estado do Rio de Janeiro

```python
df2["MÊS"] = df2["DATA FINAL"].apply(lambda x: x.month)
df_rio = df2[df2["ESTADO"] == "RIO DE JANEIRO"]
```

```python
df_month_rio = df_rio.groupby("ANO-MES")[("PREÇO MÉDIO REVENDA", "MÊS")].last()
```

```python
(df_month_rio[df_month_rio["MÊS"] == 12] / df_month_rio[df_month_rio["MÊS"] == 12].shift(1)-1) * 100
```

Criaçao de uma tabela de séria temporal com diferença absoluta e percentual entre valores mais barato e caros. Apresentando os estados com maiores e menores preços registrados

```python
df_max = df2.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"]
df_min = df2.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"]

idx_max = df2.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].idxmax()
idx_min = df2.groupby("ANO-MES")["PREÇO MÉDIO REVENDA"].idxmin()
```

```python
df_max
```

```python
df_min
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff["abs_diff"] = df_max - df_min
df_diff["percent_diff"] = (df_max - df_min) / df_min * 100
df_diff["max"] = df_max
df_diff["min"] = df_min
```

```python
df_diff
```

```python
df_diff["ESTADO_MAX"] = df2.loc[idx_max, :]["ESTADO"].values
df_diff["ESTADO_MIN"] = df2.loc[idx_min, :]["ESTADO"].values
```
